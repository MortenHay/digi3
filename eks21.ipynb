```python
from sympy import *
from sympy.physics.units import *
init_printing()
j = I
```

```python
#De givne vrdier defineres 
Urms = 231 * volt
R1 = 180 * ohm
R2 = 240 * ohm
C1 = 1 * micro*farad
L2 = 474 * milli*henry
f = 50*hertz

#Frekvens omregnes til vinkelfrekvens
w = (2*pi*f).simplify()

#Componenterne omregnes til impedanser
ZR1 = R1
ZC1 = convert_to((1/j/w/C1).simplify(),ohm)
ZR2 = R2
ZL2 = convert_to(j*w*L2.simplify(),ohm)

#Impedanserne printes
Eq(Symbol('Z_R1'),ZR1.evalf(3)),Eq(Symbol('Z_C1'),ZC1.evalf(3)),Eq(Symbol('Z_R2'),ZR2.evalf(3)),Eq(Symbol('Z_L2'),ZL2.evalf(3))
```

```python
#Erstatningsimpedansen over C1 og R1 beregnes parallelt
Z1 = (ZR1*ZC1/(ZR1+ZC1)).simplify()

#Erstatningsimpedansen over L2 og R2 beregnes serielt
Z2 = ZR2 + ZL2

#Erstatningsimpedanserne udskrives
Eq(Symbol('Z_1'),Z1.evalf(3)),Eq(Symbol('Z_2'),Z2.evalf(3))
```

```python
#Den samlede impedans beregnes parallelt
Z = (Z1*Z2/(Z1+Z2)).simplify()

#Impedansen udskrives
Z.evalf(3)
```

```python
#Den komplekse effekt beregnes som U^2/Z
S = convert_to((Urms**2/Z).simplify(),volt*ampere)

#Den komplekse effekt udskrives
Eq(Symbol('S'),S.evalf(3))
```

```python
#Vinklen p den komplekse effekt beregnes
phi = atan((im(S)/re(S)).simplify())

#Vinkeln udskrives i radianer og grader
Eq(Symbol('varphi'),(phi.evalf(3)*radian,convert_to(phi,degree).evalf(3)))
```

```python
#Effektfaktoren cos(phi) beregnes og udskrives
Eq(cos(phi).evalf(3))
```

```python
n1 = 132
n2 = 400

Rk = 0.01*ohm/kilometer
Lk = 0.2*milli*henry/kilometer
lk = 50*kilometer

phi = -acos(0.8)

Ul = 230*kilo*volt
Sl = 200*exp(-j*phi)*mega*volt*ampere

Il = conjugate(Sl/Ul).simplify()
Il.evalf()
```

```python
R = Rk*lk
L = Lk*lk
Z = R + convert_to(j*w*L,ohm)
Ut = (Ul*n1/n2).simplify()
Ut.evalf(5)
It = (Il*n2/n1).simplify()
Ut.evalf(5),It.evalf(5)
Uk = convert_to((It*Z).simplify(),volt)
Uk.evalf(5)
U = (Uk + Ut).simplify()
U.evalf(5)
```

```python
S = U*It.conjugate()
S.evalf(5)
```
